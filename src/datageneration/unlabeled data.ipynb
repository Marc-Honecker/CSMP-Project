```python
import pandas as pd
import random
```

```python
def genParam():
    # Hurst is in [0.7, 0.9]

    H = random.uniform(0.7, 0.9)
    # lambda_r is in [100nm, 300nm]
    lambda_r = random.uniform(100, 300)
    # lambda_s is in [1nm, 10nm]
    lambda_s = random.uniform(1, 10)
    # surfaceEnergy is in [1e-3, 10]
    gamma = random.uniform(1e-3, 10)
    # rho is in [1nm, 10µm]
    rho = random.uniform(0.01, 5)
    # L is in [500µm, 1100µm]
    L = random.uniform(500, 1100)

    return (H, lambda_r, lambda_s, gamma, rho, L)
```

```python
H = list()
lambda_r = list()
lambda_s = list()
gamma = list()
rho =  list()
L =  list()

# generating 5_000 samples
for i in range(5_000):
    res = genParam()

    H.append(res[0])
    lambda_r.append(res[1])
    lambda_s.append(res[2])
    rho.append(res[3])
    gamma.append(res[4])
    L.append(res[5])
```

```python
df = pd.DataFrame(data={"Hurst": H, "L":L, "lambdaR": lambda_r, "lambdaS": lambda_s, "surfaceEnergy": gamma, "rho":rho})
df.to_csv("data/RoughContact/pulloff_unlabeled.csv", sep=",", index=False)
```
